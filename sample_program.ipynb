```python
import array
import string
import sys
import random
```

```python
label_dic = {}
feature_dic = {}
feature_dic["___BIAS___"] = 0

weight_vectors = []
u_vectors = []
```

```python
class Sample:
    def __init__(self, label, features):
        self.label = label
        self.features = features
                 
def read_samples(filename, data):
    f = open(filename, 'r')
    for line in f.readlines():
        line = string.strip(line)
        items = line.split(' ')
        words = items[1:]
        label = items[0]
        if not(label_dic.has_key(label)):
            label_dic[label] = len(label_dic)
        label_id = label_dic[label]
        features = [0] # bias feature
        for word in words:
            if not(feature_dic.has_key(word)):
                feature_dic[word] = len(feature_dic)
            features.append(feature_dic[word])
        sample = Sample(label_id, features)
        data.append(sample)
#        print sample.label, sample.features

def classify(sample):
    max_score = 0
    max_i = 0
    dim = len(weight_vectors[0])
    for i in range(len(weight_vectors)):
        score = 0
        for feature_id in sample.features:
            if feature_id < dim:
                score += weight_vectors[i][feature_id]
        if score > max_score:
            max_score = score
            max_i = i
    return max_i
```

```python
training_data = []
development_data = []
read_samples("data/train.pos", training_data)

# initialization
for i in range(len(label_dic)):
    weight_vectors.append(array.array('d'))
    u_vectors.append(array.array('d'))
    for j in range(len(feature_dic)):
        weight_vectors[i].append(0)
        u_vectors[i].append(0)

# training
t = 0
for i in range(1, 5):
    num_updates = 0
    for j in range(len(training_data)):
        t += 1
        r = random.randint(0, len(training_data) - 1)
        sample = training_data[r];
        y = classify(sample)
        if (y == sample.label):
            continue
        for feature in sample.features:
            weight_vectors[sample.label][feature] += 1
            weight_vectors[y][feature] -= 1
            u_vectors[sample.label][feature] += t
            u_vectors[y][feature] -= t
        num_updates += 1
    print(i, "num_updates = ", num_updates)

# compute the average weight vector
for i in range(len(label_dic)):
    for j in range(len(weight_vectors[i])):
        weight_vectors[i][j] -= u_vectors[i][j] / (t + 1)

# evaluation
read_samples("tmp2.txt", development_data)
correct = 0
for sample in development_data:
    y = classify(sample)
    if (y == sample.label):
        correct += 1
print("accuracy = ", float(correct) / len(development_data))
```
